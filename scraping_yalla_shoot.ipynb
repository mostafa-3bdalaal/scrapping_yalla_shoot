```python
import pandas as pd
import numpy as np
import requests
import csv
from bs4 import BeautifulSoup
from itertools import zip_longest
from google.colab import files
```

```python
request=requests.get("https://yallashoot.soccer/competition/english-premier-league-2025-2026/")
request
```

```python
soup=BeautifulSoup(request.text,"lxml")
soup
```

```python
Date=[]
```

```python
matches=soup.find_all("div",{"class":"match-slim__inner-wrapper d-sm-flex"})
for i in matches:
  date=i.find("div",class_="match-slim__date d-flex d-sm-inline-block mr-1 mr-sm-0").text
  time=i.find('span',class_="match-slim__time match__time-formatted anwp-leading-1").text
  team1=i.find('div',class_="match-slim__team-home-title anwp-text-sm anwp-text-center anwp-text-sm-right anwp-break-word mr-sm-3 anwp-leading-1").text
  team2=i.find('div',class_="match-slim__team-away-title anwp-text-sm anwp-text-center anwp-text-sm-left anwp-break-word ml-sm-3 anwp-leading-1").text
  print(f"{team1} {time} {date} {team2}")
  Date.append({
    "Date":date,
    "Time":time,
    "Team1":team1,
    "Team2":team2
})
```

```python
len(matches)
```

```python
df=pd.DataFrame(Date)
df.to_excel('premier_league_tabel.xlsx',index=False)
files.download('premier_league_tabel.xlsx')
```
